```python
import pandas as pd
from langdetect import detect

df = pd.read_csv('searchlang.csv')

df['Language'] = ''

for index, row in df.head(50).iterrows():
    try:
        language = detect(row[2])
        df.loc[index, 'Language'] = language
    except:
        df.loc[index, 'Language'] = 'Not a text'

df.head()
```
